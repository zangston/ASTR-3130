```python
from astropy.table import Table
from astropy.io import fits
```

```python
%cd /home/wyz5rge/Documents/ASTR-3130/Lab 4/calibration/night1_reduced

print(fits.open('messier_3_6099763_V_1.fits')[0].header.comments['DATE-OBS'])
# dates are in UT
```

```python
%cd /home/wyz5rge/Documents/ASTR-3130/Lab 4/calibration/night1_reduced

earliest_date = None
latest_date = None

for i in range(1, 69):
    filename = 'messier_3_6099763_V_' + str(i) + '.fits'
    with fits.open(filename) as hdul:
        hdr = hdul[0].header
        obs_date = hdr.get('DATE-OBS')
        print(obs_date)
        
        # Update the earliest and latest observation dates
        if earliest_date is None or obs_date < earliest_date:
            earliest_date = obs_date
        if latest_date is None or obs_date > latest_date:
            latest_date = obs_date
    
# Print the earliest and latest observation dates
print("Earliest observation date:", earliest_date)
print("Latest observation date:", latest_date)
```
